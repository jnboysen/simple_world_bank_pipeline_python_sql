```python
import requests
import pandas as pd

##### FROM WORLD BANK API: DOWNLOAD RURAL POPULATION DATA BY COUNTRY AND YEAR ##### 
# get the url ready
rural_url = 'http://api.worldbank.org/v2/country/all/indicators/SP.RUR.TOTL/?date=2000:2020&format=json&per_page=10000'
# send the request
rural_r = requests.get(rural_url)
#rural_r.json()
rural_df = pd.DataFrame(rural_r.json()[1])


##### FROM WORLD BANK API: DOWNLOAD ELECTRICITY DATA BY COUNTRY AND YEAR ##### 

# get rural population data from the world bank
# get the url ready
elec_url = 'http://api.worldbank.org/v2/country/all/indicators/EG.ELC.ACCS.ZS/?date=2000:2020&format=json&per_page=10000'

# send the request
elec_r = requests.get(elec_url)
elec_r.json()
elec_df = pd.DataFrame(elec_r.json()[1])
```

```python
#### UNPACK DICTIONARY COLUMN FUNCTION ####

#example inputs
#df = elec_df
#unpack_columns = ["indicator","country"]

def unpack(df, unpack_columns):
    x = df.drop(columns = unpack_columns)
    for i in range(0,len(unpack_columns)):
        v = unpack_columns[i]
        y = pd.DataFrame(df[v])[v].apply(pd.Series)
        y.rename(columns={"id":('{} id'.format(v)), "value":('{} name'.format(v))}, inplace=True)
        
        x = pd.DataFrame(pd.concat([x,y], axis=1))
    return(x)
```

```python
unpack_columns = ["indicator","country"]

#unpack rural dataframe
rural_df = unpack(rural_df, unpack_columns)

#unpack electricity dataframe
elec_df = unpack(elec_df, unpack_columns)
```
